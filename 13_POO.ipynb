# Programación Orientada a Objetos (POO)

Hasta ahora escribimos código con **funciones sueltas** y **variables separadas**.
Cuando un programa crece, eso se vuelve un desorden.

La **POO** agrupa los datos y las funciones que los usan en una sola cosa llamada **objeto**.

## Clase y objeto

- **Clase** → el **molde** (la plantilla).
- **Objeto** → una **copia** creada a partir del molde (también llamada *instancia*).

Analogía: la clase es el plano de una casa, el objeto es una casa real construida con ese plano.
Con un plano puedes construir muchas casas distintas.

````python
class Perro:
    pass

# Creamos dos perros (dos objetos de la clase Perro)
firulais = Perro()
rocky = Perro()

print(type(firulais))  # <class '__main__.Perro'>
print(firulais is rocky)  # False — son objetos distintos
````

## `__init__` y `self`

- `__init__` es el **constructor** — se ejecuta automáticamente al crear el objeto.
- `self` representa **al propio objeto** que se está creando. Siempre es el primer parámetro de los métodos.

`self.nombre = nombre` guarda el valor dentro del objeto, no en una variable suelta.

````python
class Perro:
    def __init__(self, nombre, raza):
        self.nombre = nombre   # atributo de instancia
        self.raza = raza

firulais = Perro("Firulais", "Labrador")
rocky = Perro("Rocky", "Bulldog")

print(firulais.nombre, firulais.raza)  # Firulais Labrador
print(rocky.nombre, rocky.raza)        # Rocky Bulldog
````

## Métodos

Un **método** es una función definida dentro de una clase. Siempre lleva `self` como primer parámetro.
Desde un método puedes acceder a los atributos del objeto con `self.atributo`.

````python
class Perro:
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau!")

    def presentarse(self):
        print(f"Soy {self.nombre} y soy un {self.raza}")

firulais = Perro("Firulais", "Labrador")
firulais.ladrar()
firulais.presentarse()
````

## Atributos de clase vs atributos de instancia

| Tipo | Dónde se define | Lo comparten todos los objetos |
|---|---|---|
| **Instancia** | Dentro de `__init__` con `self.` | ❌ Cada objeto tiene el suyo |
| **Clase** | Directamente en el cuerpo de la clase | ✅ Lo comparten todos |

> Los atributos de clase sirven para **constantes** o **contadores** compartidos.

````python
class Perro:
    especie = "Canis familiaris"  # atributo de clase (compartido)
    cantidad = 0                  # contador de instancias

    def __init__(self, nombre):
        self.nombre = nombre       # atributo de instancia (propio)
        Perro.cantidad += 1        # aumenta el contador de la CLASE

a = Perro("Firulais")
b = Perro("Rocky")

print(a.especie, b.especie)   # misma especie
print(a.nombre, b.nombre)     # nombres distintos
print(Perro.cantidad)         # 2
````

## `__str__` — cómo se muestra el objeto

Por defecto, al imprimir un objeto sale algo feo como `<__main__.Perro object at 0x000...>`.
Definiendo `__str__` controlas cómo se ve al hacer `print(objeto)`.

````python
class Perro:
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    def __str__(self):
        return f"Perro({self.nombre}, {self.raza})"

firulais = Perro("Firulais", "Labrador")
print(firulais)  # Perro(Firulais, Labrador)
````

## Herencia — reutilizar una clase

Una clase puede **heredar** de otra: recibe todos sus atributos y métodos, y puede añadir o modificar lo que necesite.

Sintaxis: `class Hija(Padre):`

````python
class Animal:
    def __init__(self, nombre):
        self.nombre = nombre

    def hablar(self):
        print(f"{self.nombre} hace un sonido")

class Perro(Animal):
    def hablar(self):   # sobrescribimos el método
        print(f"{self.nombre} dice: ¡Guau!")

class Gato(Animal):
    def hablar(self):
        print(f"{self.nombre} dice: Miau")

for a in [Perro("Firulais"), Gato("Michi"), Animal("Pato")]:
    a.hablar()
````

## `super()` — usar el método del padre

Si la clase hija necesita **más atributos** que el padre, usa `super().__init__(...)` para reutilizar el constructor del padre en vez de repetirlo.

````python
class Animal:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

class Perro(Animal):
    def __init__(self, nombre, edad, raza):
        super().__init__(nombre, edad)  # reutiliza Animal.__init__
        self.raza = raza

p = Perro("Firulais", 3, "Labrador")
print(p.nombre, p.edad, p.raza)
````

## Encapsulación — atributos "privados"

Python no tiene atributos realmente privados, pero usa **convenciones**:

| Nombre | Significado |
|---|---|
| `self.nombre` | Público, cualquiera puede tocarlo |
| `self._saldo` | "No tocar desde fuera" (convención, sigue siendo accesible) |
| `self.__password` | Name mangling — Python renombra el atributo para dificultar el acceso |

> Regla práctica: si un atributo solo debe modificarse mediante métodos (ej: `saldo` de una cuenta), usa `_` delante.

````python
class CuentaBancaria:
    def __init__(self, titular, saldo_inicial=0):
        self.titular = titular
        self._saldo = saldo_inicial   # convención: no tocar desde fuera

    def depositar(self, monto):
        if monto <= 0:
            print("Monto inválido")
            return
        self._saldo += monto

    def retirar(self, monto):
        if monto > self._saldo:
            print("Saldo insuficiente")
            return
        self._saldo -= monto

    def consultar(self):
        return self._saldo

c = CuentaBancaria("Lian", 100)
c.depositar(50)
c.retirar(30)
print(c.consultar())  # 120
````

## Caso especial: métodos dunder (`__nombre__`)

Son métodos con doble guión bajo que Python llama automáticamente en ciertas situaciones:

| Método | Se llama cuando... |
|---|---|
| `__init__` | Se crea el objeto |
| `__str__` | Se hace `print(obj)` o `str(obj)` |
| `__len__` | Se hace `len(obj)` |
| `__eq__` | Se comparan objetos con `==` |

````python
class Carrito:
    def __init__(self):
        self.productos = []

    def agregar(self, producto):
        self.productos.append(producto)

    def __len__(self):
        return len(self.productos)

    def __str__(self):
        return f"Carrito con {len(self)} productos: {self.productos}"

c = Carrito()
c.agregar("pan")
c.agregar("leche")
print(len(c))  # 2
print(c)       # Carrito con 2 productos: ['pan', 'leche']
````

## Caso especial: olvidar `self`

Error muy común al empezar — si omites `self`, Python lanza `TypeError`.

```python
class Perro:
    def ladrar():          # ❌ falta self
        print("Guau")

Perro().ladrar()           # TypeError: ladrar() takes 0 positional arguments but 1 was given
```

Python siempre pasa el objeto como primer argumento automáticamente. Si no lo recibes con `self`, se rompe.

## Caso especial: atributo de clase mutable

⚠️ Si pones una **lista** como atributo de clase, **todos los objetos la comparten**.
Es un bug muy frecuente.

````python
# ❌ MAL — la lista está en la CLASE, no en cada instancia
class Mochila:
    items = []   # compartida entre todos

    def agregar(self, item):
        self.items.append(item)

a = Mochila()
b = Mochila()
a.agregar("libro")
print(b.items)   # ¡['libro']! b heredó la lista de a

# ✅ BIEN — cada objeto crea su propia lista en __init__
class MochilaOK:
    def __init__(self):
        self.items = []

    def agregar(self, item):
        self.items.append(item)

x = MochilaOK()
y = MochilaOK()
x.agregar("libro")
print(y.items)   # []
````
