6) El Traductor de "camelCase" a "snake_case"
En muchos lenguajes de programación se usa camelCase (ej. nombreDeUsuario), pero en Python preferimos snake_case (ej. nombre_de_usuario)

- Reto: Escribe un programa que pida al usuario el nombre de una variable en camelCase y lo devuelva convertido a snake_case.

- Nivel Pro: Deberás iterar sobre cada carácter de la cadena, detectar cuáles son mayúsculas y reemplazarlas por un guion bajo seguido de la letra en minúscula

```python
# Definimos la funcion y que pida un str
def convertir(text):
  # Que el primer caracter se muestre en minusculas
  result = text[0].lower()
  # Por cada caracter del texto a partir del 1 hasta el final
  for char in text[1:]:
    # Si el caracter esta en mayusculas agrega un guion bajo "_" y agregale el mismo caracter seleccionado
    if char.isupper():
      result += "_" + char.lower()
    # Si no sigue con el siguiente caracter
    else:
      result += char
  # Devuelve el resultado en un "return" para que se pueda usar luego
  return result

name = input("camelCase: ")
print(convertir(name))
```

```python
# Pide un texto en mayusculas tipo CAMELcASE
def convertir(text):
  result = text[0].upper()
  for char in text[1:]:
    if char.islower():
      result += "_" + char.upper()
    else:
      result += char
  return result

name = input("")# HOLAmUNDO
print(convertir(name))
```

11) El Creador de Siglas Académicas
Escribe un programa que reciba el nombre de una institución de __dos palabras__ (ej. "harvard university") y devuelva sus siglas en mayúsculas y separadas por puntos.
- Reto: Debes convertir la entrada a un formato donde obtengas "H.U.".

- Herramientas sugeridas: split(), slicing para obtener la primera letra de cada palabra y f-strings para el formato final

```python
nombre_institucion = input("Ingrese el nombre de la Institucion: ").split(" ")
primera_sigla = ((nombre_institucion[0])[0]).upper()
segunda_sigla = ((nombre_institucion[1])[0]).upper()

print(f"{primera_sigla}.{segunda_sigla}.")
```

13. Extractor de Dominios de Correo
Solicita al usuario su correo electrónico (ej. "usuario@dominio.com") y muestra únicamente el nombre del dominio, sin el nombre de usuario ni el punto del final (ej. "dominio").
- Reto: No puedes usar bucles para buscar la @. Debes "romper" la cadena de forma quirúrgica.
- Herramientas sugeridas: Usar split("@") para separar el usuario, y luego volver a usar split(".") sobre el resultado de la derecha para obtener solo la parte central

```python
# Primera version
correo = input("Ingrese un correo: ")
separar = correo.split("@")
dominio = separar[1].replace(".com", "")
print(dominio)
```

```python
# Segunda version
correo = input("Ingrese un correo: ")
# El correo cuando encuentre "@" se va a separar y almacenarse en una lista: ['usuario', 'separado']
usuario, separado = correo.split("@")
# usuario: lian | separado: dev.com
# separado vamos a volver a separarlo cuando encuentre "." y resultara en ['dev', 'com'], con esto llamamos al dato en el primer orden "[0]" que es 'dev'
dominio = separado.split(".")[0]
print(dominio)
```
