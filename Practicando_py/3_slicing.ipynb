## Slicing

1. Primeras y últimas letras

Pide una palabra al usuario e imprime su primera letra, su última letra, y las 3 letras del medio (desde índice 1 hasta -1).

```python
palabra  = input("")
p1 = palabra[0]
p2 = palabra[-1]
p3 = palabra[1:-1]
print(f"{p1}\n{p2}\n{p3}")
```

2. ¿Es palíndromo?
   
Pide una palabra y compara si es igual a sí misma al revés usando slicing. Muestra "Es palíndromo" o "No es palíndromo".

```python
palabra = input("")
palindromo = palabra [::-1]
if palabra == palindromo:
  print(f"Es palindromo de {palabra} ---> {palindromo}")
else:
  print(f"No es palindromo de {palabra} --->X {palindromo}")
```

3. Censurar una palabra
   
Pide una palabra y muéstrala con solo la primera y última letra visibles, reemplazando el medio con asteriscos. Usa slicing para obtener las letras y concatena con "*".

```python
palabra = input("")
asteriscos = "*" * (len(palabra) - 2)
censura = palabra[0] + asteriscos + palabra[-1]
print(censura)
```

4. Nombre abreviado
   
Pide al usuario su nombre completo (ej. "Juan Perez") y muestra solo la inicial del nombre seguida de un punto y el apellido completo.

```python
usuario = input("Ingrese un nombre:").strip().title()
nombre, apellido = usuario.split()

print(f"{nombre[0]}. {apellido}")
```

5. Validar extensión de archivo
   
Pide el nombre de un archivo y verifica si termina en ".py" usando slicing con índice negativo. Muestra "Es un archivo Python" o "No es un archivo Python".

```python
nombre_archivo = input("Ingrese el nombre del archivo: ")

if nombre_archivo[-3:] == ".py":
  print(f"{nombre_archivo} | Es un archivo Python")
else:
  print(f"{nombre_archivo} | No es un archivo Python")
```

6. Detector de dominio de correo
   
Pide un correo electrónico y determina si es Gmail, Hotmail o Yahoo. Si no es ninguno, muestra "Dominio desconocido".

Ejemplo: lian@gmail.com → Gmail detectado

```python
correo = input("Ingrese un correo @ Gmail, Hotmail o Yahoo ").strip().lower()

if correo[-9:] == "gmail.com":
  print("Gmail detectado")
elif correo[-11:] == "hotmail.com":
  print("Hotmail detectado")
elif correo[-9:] == "yahoo.com":
  print("Yahoo detectado")
else:
  print("Dominio desconocido")
```

7. Título de película abreviado
   
Pide el título de una película y muestra solo las primeras 10 letras seguidas de "..." si el título tiene más de 10 caracteres. Si tiene 10 o menos, muéstralo completo.

Ejemplo: "El Señor de los Anillos" → "El Señor d..."

```python
pelicula = input("Ingresa el nombre de una pelicula: ")

if len(pelicula) > 10:
  print(f"{pelicula[:10]}...")
else:
  print(pelicula)
```

8. PIN oculto
   
Pide al usuario un PIN numérico de 4 dígitos y muéstralo con los primeros 2 dígitos visibles y los últimos 2 reemplazados por "**".

Ejemplo: "1984" → "19**"

```python
PIN = input("Ingrese el PIN: ")
mostrar = PIN[:2]
if len(PIN) > 4:
  print("---El PIN debe tener solo 4 digitos---")
else:
  print(f"{mostrar}**")
```

9. Invertir solo el apellido

Pide nombre y apellido. Muestra el nombre normal pero el apellido al revés.

Ejemplo: "Lian Corper" → "Lian reproC"

```python
usuario = input("Ingrese el nombre completo: ").strip().title()
nombre, apellido = usuario.split()
usuario_reves = f"{nombre} {apellido[::-1]}"
print(usuario_reves)
```

10. Verificador de formato de fecha

Pide una fecha en formato DD/MM/YYYY y extrae con slicing el día, mes y año por separado. Muestra cada uno en una línea distinta.

Ejemplo: "24/03/2026" → Día: 24 | Mes: 03 | Año: 2026

```python
fecha = input("Ingrese una fecha en formato DD/MM/YYYY: ")
dia = fecha[0:2]
mes = fecha[3:5]
año = fecha[6:10]

print(f"Dia: {dia} | Mes: {mes} | Año: {año}")
```
