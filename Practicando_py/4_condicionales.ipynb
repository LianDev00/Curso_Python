1. El Pensamiento Profundo (Deep Thought)
Implementa un programa que haga la "Gran Pregunta sobre la Vida, el Universo y Todo lo Demás"

Reto: Si el usuario introduce "42", "forty-two" o "forty two", el programa debe imprimir "Yes"
. De lo contrario, debe imprimir "No"

Requisito: La validación debe ser insensible a mayúsculas y debe ignorar espacios adicionales al inicio o al final

```python
respuesta = input("Cual es la Gran Pregunta sobre la Vida, el Universo y Todo lo Demás?: ").strip().lower().replace("-", " ")

if respuesta == "42" or respuesta == "forty two":
  print("yes")
else:
  print("no")
```

3. Detector de Extensiones de Archivos
Los navegadores usan tipos de medios (MIME types) para saber cómo mostrar archivos

Reto: Pide al usuario el nombre de un archivo (ej. documento.pdf).
Si termina en .gif, imprime image/gif

Si termina en .jpg o .jpeg, imprime image/jpeg

Si termina en .png, imprime image/png

Si termina en .pdf, imprime application/pdf

Para cualquier otra extensión o si no tiene, imprime application/octet-stream

Requisito: Usa condicionales para evaluar el final de la cadena de texto de manera insensible a mayúsculas

```python
nombre = input("Ingresa el nombre del archivo y su extension: ").strip().lower().split(".")
if nombre[1] == "jpg" or nombre[1] == "jpeg":
  print("image/jpeg")
elif nombre[1] == "png":
  print("image/png")
elif nombre[1] == "pdf":
  print("application/pdf")
elif nombre[1] == "docx":
  print("application/word")
elif nombre[1] == "xls":
  print("application/excel")
else:
  print("application/octet-stream")
```

4. El Intérprete Matemático
Crea un programa que permita a los usuarios hacer cálculos básicos sin saber programar

Reto: Pide una expresión aritmética en formato x y z (ej. 1 + 1), donde x y z son números enteros e y es un operador (+, -, *, /)

Lógica: Usa el método split para separar la cadena en tres partes
. Luego, usa condicionales para identificar el operador y realizar la operación matemática correspondiente

Salida: Muestra el resultado como un número de punto flotante (float) formateado a un decimal

```python
calculo = input('Ingrese un calculo "+, -, *, /": ').strip().split()
calculo [0] = int(calculo [0])
calculo [2] = int(calculo [2])
if calculo[1] == "+":
  resultado = calculo[0] + calculo [2]
  print(f"{resultado:.1f}")
elif calculo[1] == "-":
  resultado = calculo[0] - calculo [2]
  print(f"{resultado:.1f}")
elif calculo[1] == "*":
  resultado = calculo[0] * calculo [2]
  print(f"{resultado:.1f}")
elif calculo[1] == "/":
  resultado = calculo[0] / calculo [2]
  print(f"{resultado:.1f}")
else:
  print("Ingrese un operador correcto")
```

5. Validador de Matrículas (Vanity Plates)
En Massachusetts, puedes elegir letras y números para tu matrícula, pero hay reglas estrictas

Reto: Pide al usuario una matrícula y determina si es "Valid" o "Invalid"

Reglas a verificar con condicionales:
Debe empezar con al menos dos letras

Debe tener entre 2 y 6 caracteres

Los números no pueden estar en medio; deben ir al final (ej. AAA222 es válido, AAA22A no lo es)

El primer número utilizado no puede ser un '0'

No se permiten puntos, espacios ni signos de puntuación

Pista de diseño: Para el ejercicio 5, recuerda que puedes usar s[0:2] para obtener los primeros caracteres (slicing) y métodos como .isalpha() o .isdigit() para tus condiciones.

```python
placa = input("Placa: ").strip().upper()

# Se debe ingresar una placa de entre 2 a 6 caracteres
if len(placa) < 2 or len(placa) > 6:
  print(f"{placa} INVALIDO: La placa debe tener de 2 a 6 caracteres")
# Debe empezar si o si con 2 letras
elif not placa[0].isalpha() or not placa[1].isalpha():
  print(f"{placa} INVALIDO: Se debe empezar con al menos 2 letras")
# No debe haber signos ni espacios
elif not placa.isalnum():
  print(f"{placa} INVALIDO: No debe tener signos ni espacios")
# El primer numero no puede ser 0
elif len(placa) > 2 and placa[2].isdigit() and placa[2] == '0':
  print(f"{placa} INVALIDO: El primer numero no puede ser 0")
elif len(placa) > 3 and placa[3].isdigit() and placa[2].isdigit() and placa[3] == '0':
# Las letras no pueden ir despues de un numero
  print(f"{placa} INVALIDO")
elif len(placa) > 3 and placa[2].isdigit() and placa[3].isalpha():
    print(f"{placa} INVALIDO")
elif len(placa) > 4 and placa[3].isdigit() and placa[4].isalpha():
    print(f"{placa} INVALIDO")
elif len(placa) > 5 and placa[4].isdigit() and placa[5].isalpha():
    print(f"{placa} INVALIDO")

else:
    print(f"{placa} VALIDO")
```

  6. Clasificador de IMC
  
Pide al usuario su peso (kg) y altura (m). Calcula el IMC (peso / altura²) y muestra la categoría: Bajo peso (< 18.5), Normal (18.5–24.9), Sobrepeso (25–29.9), Obesidad (≥ 30).

```python
peso = float(input("Ingrese su peso en Kg: "))
altura = float(input("Ingrese su altura en metros:"))
calculo_imc = peso / (altura ** 2)

if calculo_imc <= 18.5:
  print(f"{calculo_imc:.2f}: Bajo")
elif calculo_imc <= 24.9:
  print(f"{calculo_imc:.2f}: Normal")
elif calculo_imc <= 29.9:
  print(f"{calculo_imc:.2f}: Sobrepeso")
else:
  print(f"{calculo_imc:.2f}: Obesidad")
```

7. Calculadora de descuentos
Eres cajero en una tienda y debes aplicar descuentos según la edad del cliente.

Reto: Pide el precio de un producto y la edad del cliente, aplica el descuento correspondiente y muestra el precio final.
Reglas:

Menores de 12 años: 20% de descuento
Mayores de 60 años: 15% de descuento
Entre 12 y 59 años: sin descuento

Pista de diseño: Recuerda que puedes calcular el descuento así: precio * 0.20 y el precio final como precio - descuento. Muestra el resultado con 2 decimales.

```python
edad = int(input("Edad del cliente: "))
precio = float(input("Precio del producto: "))

if edad < 12:
  descuento = precio * 0.20
elif edad >= 60:
  descuento = precio * 0.15
else:
  descuento = 0

print(f"Al tener: \t\t\t{edad} años \nObtienes un descuento de: \t${descuento:.2f} \nA tu producto de: \t\t${precio:.2f} \nPrecio Final: \t\t\t${(precio - descuento):.2f}")
```

8. Piedra, Papel o Tijera
El clásico juego de manos, ahora en Python.

Reto: Pide al usuario su elección y la del rival. Determina quién ganó.

Reglas:
Piedra le gana a Tijera |
Tijera le gana a Papel |
Papel le gana a Piedra |
Si ambos eligen igual, es empate |
Si alguna entrada no es válida, muestra "Entrada inválida"

Pista de diseño: Puedes usar .lower() para evitar problemas con mayúsculas. Piensa en cuántas combinaciones posibles existen y cómo organizarlas con if/elif/else.

9. Conversor de Notas

Eres profesor y necesitas convertir notas numéricas a letras para el boletín de calificaciones.

Reto: Pide una nota numérica del 0 al 100 y muestra su equivalente en letra.

Reglas: 
- 90 – 100 → A
- 80 – 89 → B
- 70 – 79 → C
- 60 – 69 → D
- 0 – 59 → F
- Fuera de rango → "Nota inválida"

Pista de diseño: Recuerda que en elif no necesitas repetir la parte izquierda del rango, Python ya sabe que las condiciones anteriores fallaron.

```python
nota = int(input("Ingrese la nota: "))

if nota > 100 or nota < 0:
  categoria = "---NOTA INVALIDA---"
elif nota >= 90:
  categoria = "A"
elif nota >= 80:
  categoria = "B"
elif nota >= 70:
  categoria = "C"
elif nota >= 60:
  categoria = "D"
else:
  categoria = "F"

print(f"Nota: \t\t{nota:.2f} \nCategoria: \t{categoria}")
```
