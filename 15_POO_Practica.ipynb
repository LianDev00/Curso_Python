### 10. `Banco` con múltiples cuentas (integrador)

Pone todo junto. Crea:

**`Cuenta`** con `titular`, `_saldo`, métodos `depositar` y `retirar` que **lanzan** `ValueError` en casos inválidos (como ej. 4).

**`Banco`** con:
- `cuentas` → dict `{titular: Cuenta}`
- `crear_cuenta(titular, saldo=0)` — si el titular ya existe, `raise ValueError("Ya existe")`
- `transferir(origen, destino, monto)`:
  - Si alguna cuenta no existe → `raise KeyError`
  - Retira de `origen` (puede lanzar ValueError si no hay saldo)
  - Si el retiro falla, **NO** deposita en destino (atomicidad)
  - Si todo OK, deposita en `destino`
- `__str__` → muestra todas las cuentas con sus saldos

Simula: crear 3 cuentas, hacer transferencias (una exitosa, una con saldo insuficiente, una con cuenta inexistente) envueltas en `try/except`.

**Reto:** La transferencia debe ser **atómica** — si falla el retiro, NO se deposita. El `try/except` envuelve solo el retiro; si se dispara, `return` sin depositar.

**Pista:**
```python
try:
    self.cuentas[origen].retirar(monto)
except ValueError:
    raise  # re-lanza el error al código que llamó
self.cuentas[destino].depositar(monto)
```
