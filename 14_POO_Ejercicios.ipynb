## Fáciles

### 1. Clase `Persona`

Crea una clase `Persona` con atributos `nombre` y `edad`, y un método `saludar()` que imprima:

```
Hola, soy Lian y tengo 26 años.
```

Crea dos personas distintas y llama a `saludar()` en cada una.

**Reto:** Usa f-string dentro del método `saludar`.

**Pista:** Recuerda `__init__(self, nombre, edad)` y `self.nombre = nombre`.

### 2. Clase `Rectangulo`

Crea una clase `Rectangulo` con atributos `ancho` y `alto`. Añade dos métodos:
- `area()` → devuelve `ancho * alto`
- `perimetro()` → devuelve `2 * (ancho + alto)`

Crea un rectángulo 5×3 y muestra ambos valores.

**Resultado esperado:**
```
Área: 15
Perímetro: 16
```

**Reto:** Los métodos deben **devolver** (`return`), no imprimir.

**Pista:** `r.area()` llama al método — con paréntesis, no sin ellos.

### 3. Clase `Libro` con `__str__`

Crea una clase `Libro` con `titulo`, `autor` y `paginas`. Define `__str__` para que al imprimir el libro salga:

```
'Cien años de soledad' de Gabriel García Márquez (471 páginas)
```

**Reto:** Usa comillas simples alrededor del título dentro del f-string.

**Pista:** `__str__` debe **devolver** un string, no imprimirlo.

### 4. Clase `Contador`

Crea una clase `Contador` que empieza en 0 y tiene tres métodos:
- `incrementar()` → suma 1
- `decrementar()` → resta 1
- `valor()` → devuelve el valor actual

Prueba haciendo 3 incrementos y 1 decremento. Debe mostrar `2`.

**Reto:** El valor inicial debe guardarse como atributo dentro de `__init__`, no como variable global.

**Pista:** `self.valor_actual = 0` en `__init__` y luego `self.valor_actual += 1`.

## Intermedios

### 5. `CuentaBancaria` con validación

Crea una clase `CuentaBancaria` con `titular` y `_saldo` (inicial 0). Implementa:
- `depositar(monto)` → suma al saldo si `monto > 0`, si no imprime `"Monto inválido"`
- `retirar(monto)` → resta si hay saldo suficiente, si no imprime `"Saldo insuficiente"`
- `__str__` → devuelve `"Cuenta de Lian: $120"`

**Resultado esperado** (tras depositar 100 y retirar 30):
```
Cuenta de Lian: $70
```

**Reto:** Prueba también casos inválidos (retirar más del saldo, depositar -10).

**Pista:** Usa `_saldo` (con guión bajo) por convención para indicar que no debe tocarse directamente.

### 6. `Carrito` de compras

Crea una clase `Carrito` que guarda productos como lista de tuplas `(nombre, precio)`. Implementa:
- `agregar(nombre, precio)` → añade el producto
- `total()` → devuelve la suma de los precios
- `__len__` → cantidad de productos
- `__str__` → muestra los productos línea por línea y el total al final

**Resultado esperado:**
```
- pan: $2.5
- leche: $4.0
- queso: $8.0
Total: $14.5 (3 productos)
```

**Reto:** Usa `len(self)` dentro de `__str__` para no repetir lógica.

**Pista:** `sum(p for n, p in self.productos)` saca el total con una comprehension.

### 7. Contador de instancias

Crea una clase `Usuario` con atributo de instancia `nombre` y un **atributo de clase** `total_usuarios` que cuenta cuántos usuarios se han creado.

Crea 4 usuarios y al final imprime: `Total de usuarios registrados: 4`.

**Reto:** `total_usuarios` NO debe ir dentro de `__init__` — va en el cuerpo de la clase.

**Pista:** En `__init__` haz `Usuario.total_usuarios += 1`. Accede al final con `Usuario.total_usuarios` (no `self`).

## Difíciles

### 8. Herencia: `Animal`, `Perro`, `Gato`

Crea una clase base `Animal` con `nombre`, `edad` y un método `hablar()` que imprima `"[nombre] hace un sonido"`.

Luego crea dos subclases:
- `Perro(Animal)` — sobrescribe `hablar()` para que imprima `"[nombre] dice: ¡Guau!"`
- `Gato(Animal)` — añade el atributo `color` (usa `super().__init__`) y sobrescribe `hablar()` para decir `"Miau"`

Crea una lista con 3 animales mezclados y recórrela llamando a `hablar()` en cada uno.

**Reto:** El `for` no debe preguntar de qué tipo es cada animal — el polimorfismo hace el trabajo.

**Pista:** `Gato.__init__` debe recibir `nombre, edad, color`, llamar a `super().__init__(nombre, edad)` y luego guardar `self.color = color`.

### 9. Biblioteca con préstamos

Crea dos clases:

**`Libro`** con `titulo`, `autor` y `disponible` (inicial `True`).

**`Biblioteca`** con una lista de libros y métodos:
- `agregar(libro)` → añade un libro a la biblioteca
- `prestar(titulo)` → si el libro existe y está disponible, lo marca como `disponible=False` e imprime `"Prestado: [titulo]"`. Si no está disponible o no existe, mensaje de error
- `devolver(titulo)` → lo marca como disponible de nuevo
- `listar()` → imprime todos los libros con su estado `(disponible)` o `(prestado)`

**Reto:** No permitas prestar un libro que ya está prestado. Busca por título case-insensitive.

**Pista:** Dentro de `prestar` recorre `self.libros` con un `for` y compara `libro.titulo.lower() == titulo.lower()`. Usa `break` al encontrarlo.

### 10. Jugador RPG con combate

Crea una clase `Jugador` con:
- `nombre`
- `_vida` (inicial 100) — privado por convención
- `_ataque` (inicial 20)
- `_defensa` (inicial 10)

Métodos:
- `recibir_daño(daño)` → resta `daño - self._defensa` de la vida (mínimo 0 de daño, no puede curar al recibir 5 daño con 10 defensa). Si la vida llega a 0 o menos, imprime `"[nombre] ha caído"` y pone `_vida = 0`
- `atacar(otro)` → llama a `otro.recibir_daño(self._ataque)` e imprime `"[atacante] ataca a [otro] con [ataque] de daño"`
- `esta_vivo()` → devuelve `True` si `_vida > 0`
- `__str__` → `"Lian - Vida: 85/100"` (requiere guardar la vida máxima también)

Simula un combate por turnos entre dos jugadores hasta que uno caiga. Usa un `while` que avanza mientras ambos estén vivos.

**Reto:** El daño recibido nunca debe ser negativo — si la defensa es mayor que el ataque, el daño final es 0.

**Pista:**
- Guarda `self._vida_max = vida` al inicio para mostrarla en `__str__`
- Usa `max(0, daño - self._defensa)` para evitar daño negativo
- En el bucle de combate, alterna turnos con dos líneas: `a.atacar(b)` y `b.atacar(a)`, verificando `esta_vivo()` entre cada ataque
