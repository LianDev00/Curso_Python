```python
# Funcion que llama a otra funcion
def es_par(n):
  return n % 2 == 0

def filtrar_pares(lista):
  resultado = []  # Iniciamos una lista vacia para poder acumularlo
  for n in lista: # Recorremos la lista por cada numero
    if es_par(n): # Si el numero es par...
      resultado.append(n) # Lo agregamos a la lista "resultado"
  return resultado # Regresamos la lista como un valor

numeros = [1, 2, 3, 4, 5, 6, 7, 8, 9, 0]
print(filtrar_pares(numeros))
```

🎯 Reto

💡 Pista de diseño
Usa return imc, categoria para devolver dos valores a la vez. Al llamar la función, desempácalos: imc, cat = calcular_imc(peso, altura). Recuerda: IMC = peso / altura².

```python
def main()  :
  peso = int(input("Ingrese su peso (kg): "))
  altura = float(input("Ingrese su altura (m): "))
  resultado = calcular_imc(peso, altura)
  print(f"Su IMC es: {resultado:.2f}")
  
def calcular_imc(peso, altura):
  return peso / (altura ** 2)

main()
```

```python
def calcular_imc(peso, altura):
  imc = peso / (altura ** 2)
  
  if imc < 18.5:
    categoria = "Bajo peso"
  elif imc < 25:
    categoria = "Normal"
  elif imc < 30:
    categoria = "Sobrepeso"
  else:
    categoria = "Obesidad"
  return imc, categoria

def main():
  peso = float(input("Peso (Kg): "))
  altura = float(input("Altura (m): "))
  imc, categoria = calcular_imc(peso, altura)
  print(f"IMC: {imc:.2f} -- {categoria}")
  
main()
```

2. Conversor de Temperaturas

Crea tres funciones: celsius_a_fahrenheit(c), fahrenheit_a_celsius(f) y main(). El programa debe preguntar al usuario qué conversión quiere hacer y en qué dirección.

🎯 Reto
Cada función de conversión solo hace el cálculo y retorna el resultado — no imprime nada. Todo el input/output se maneja en main(). Usa un menú con while True para repetir hasta que el usuario elija salir.

💡 Pista de diseño
Fórmulas: F = (C × 9/5) + 32 y C = (F - 32) × 5/9. Usa input().strip().lower() para normalizar la opción del usuario.

```python
def celsius_a_fahrenheit(c):
  return (c * 9 / 5)+ 32
  
def fahrenheit_a_celsius(f):
  return (f - 32) * 5 / 9

def main():
  while True:
    opcion = input("1. Celsius a Fahrenheit \n2. Fahrenheit a Celsius \n3. Salir").strip().lower()
    
    if opcion == "1":
      c = float(input("Ingrese la temperatura Celsius: "))
      resultado = celsius_a_fahrenheit(c)
      print(f"{c}°C --> {resultado:.1f}°F")
    elif opcion == "2":
      f = float(input("Ingrese la temperatura fahrenheit: "))
      resultado = fahrenheit_a_celsius(f)
      print(f"{f}°F --> {resultado:.1f}°C")
    elif opcion == "3":
      print("Saliendo....")
      break
    
main()
```

Crea una función es_palindromo(texto) que reciba una cadena y retorne True si es palíndromo o False si no lo es. Debe ignorar mayúsculas y espacios.

🎯 Reto
La función debe limpiar el texto antes de comparar: quitar espacios y convertir a minúsculas. No uses [::-1] directamente — construye la comparación limpiando primero el texto.

💡 Pista de diseño
Usa texto.lower().replace(" ", "") para normalizar antes de comparar con [::-1]. La función solo retorna True o False — el mensaje lo maneja main().

```python
def es_palindromo(texto):
  if texto == texto[::-1]:
    return True, "Es palindromo"
  else: 
    return False, "No es palindromo"
  
def main():
  palin = input("Ingrese una palabra: ").strip().lower().replace(" ", "")
  ok, comparar = es_palindromo(palin)
  print(f"{palin} --> {comparar}")
  
main()
```

```python
def es_palindromo(texto):
  # El texto se limpia solo en esta seccion, pero se sigue mostrando igual al momento de devolverlo en la consola
  limpio = texto.lower().replace(" ", "") 
  return limpio == limpio[::-1] # True o False

def main():
  texto = input("Ingresa una palabra: ").strip()
  if es_palindromo(texto):
    print(f'"{texto}" ES Palindromo')
  else:
    print(f'"{texto}" NO es Palindromo')
    
main()
```

4. Generador de Contraseña Segura

Crea una función validar_password(password) que verifique si una contraseña es segura y retorne una lista con los errores encontrados. Si la lista está vacía, la contraseña es válida.

🎯 Reto
La función debe verificar al menos 4 reglas: mínimo 8 caracteres, al menos una mayúscula, al menos un número, sin espacios. Retorna una lista de strings con cada error, o una lista vacía si todo está bien.

💡 Pista de diseño
Crea una lista errores = [] dentro de la función. Por cada regla que falle, haz errores.append("descripción del error"). Al final retorna errores. En main(), si len(errores) == 0 → válida.

```python
def validar_password(password):
  if len(password) < 8:
    return "Debe tener al menos 8 caracteres"
  if " " in password:
    return "No debe tener espacios"
  if not any(c.isupper() for c in password):
    return "Debe tener al menos una mayuscula"
  if not any(c.isdigit() for c in password):
    return "Debe tener al menos un numero"
  return "Acceso Valido"

def main():
  passw = input("Ingrese la contraseña: ").strip()
  resultado = validar_password(passw)
  print(resultado)
  
main()
```

```python
def validar_password(password):
  errores = []
  if len(password) < 8:
    errores.append("Debe tener al menos 8 caracteres")
  if " " in password:
    errores.append("No debe tener espacios")
  if not any(c.isupper() for c in password):
    errores.append("Debe tener al menos una mayuscula")
  if not any(c.isdigit() for c in password):
    errores.append("Debe tener al menos un numero")
  return errores

def main():
  password = input("Ingrese la contraseña: ").strip()
  errores = validar_password(password)
  if len(errores) == 0:
    print("--- Acceso Valido ---")
  else:
    print("Errores encontrados:")
    for e in errores:
      print(f"\t- {e}")
    print("Saliendo....")  
    
main()
```

5. Cifrado César con Funciones

Refactoriza el cifrado César que hiciste antes en funciones. Crea cifrar(texto, n) que cifre el texto y descifrar(texto, n) que lo descifre. Ambas deben usar una función auxiliar interna.

🎯 Reto
Crea una tercera función desplazar(texto, n) que haga el trabajo real. cifrar() la llama con +n y descifrar() la llama con -n. Así evitas duplicar código.

💡 Pista de diseño
La función desplazar(texto, n) recorre cada carácter: si es letra usa chr(ord(c) + n), si no lo deja igual. cifrar y descifrar son solo envoltorios que llaman a desplazar con signo positivo o negativo.

```python
def desplazar(texto, n):
  resultado = ""
  for c in texto:
    if c.isalpha():
      resultado += chr(ord(c) + n)
    else:
      resultado += c
  return resultado
  
def cifrar(texto, n):  
  return desplazar(texto, n)
  
def descifrar(texto, n):
  return desplazar(texto, -n)

def main():
  texto = input("Ingresa un texto: ").lower()
  n = int(input("Ingresa el desplazamiento"))
  cifrado = cifrar(texto, n)
  descifrado = descifrar(cifrado, n)
  print(f"Original: {texto}")
  print(f"Cifrado: {cifrado}")
  print(f"Descifrado: {descifrado}")
  
main()
```
