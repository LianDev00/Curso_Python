1. Imprime los números del 1 al 10, uno por línea, usando un bucle for.

### Bucles while

3. Partiendo de 10, imprime cada número hasta llegar a 1 usando while. Al final imprime "¡Despegue!".
   
    Restricción: solo while · sin range()

```python
numero = 10

while True:
  print (numero)
  numero = numero - 1
  if numero == 0:
    print("¡Despegue!")
    break
```

4. Empieza con n = 1. Mientras n sea menor que 100, duplícalo e imprímelo. ¿Cuántas veces se duplicó?

    Restricción: usa while · imprime el conteo al final

```python
n = 1

while n < 100:
  n *= 2
  print(n)
```

6. Dado un valor de temperatura hardcodeado, clasifícala:

    - menos de 0 → "helada"
    - 0–15 → "fría"
    - 16–30 → "agradable"
    - más de 30 → "calurosa"

    Restricción: usa if / elif / else

### Listas con .append()

9. Empieza con una lista vacía. Usando un for del 1 al 5, agrega el cuadrado de cada número a la lista. Imprime la lista final.

    Restricción: usa .append() · sin list comprehension

```python
lista = []

for n in range(1, 6):
  cuad = n ** 2
  lista.append(cuad)

print(lista)
```

10. Dada la lista [-3, 5, -1, 8, 0, 2], crea una nueva lista que solo contenga los números positivos usando append.
    
    Restricción: usa for, if y .append()· sin filter()

```python
lista = [-3,5,-1,8,0,2]
nl = []

for l in lista:
  if l > 0:
    nl.append(l)
    
print(nl)
```

### Diccionarios con .get()

11. Tienes el diccionario {"manzana": 1.5, "pan": 2.0, "leche": 3.2}. Busca el precio de "pan" y de "arroz" usando .get(). Si no existe, muestra "no disponible".
    
    Restricción: usa .get() con valor por defecto · sin if

```python
frutas = {
          "manzana": 1.5, 
          "pan": 2.0, 
          "leche": 3.2
}

pan = frutas.get("pan", "no disponible")
arroz = frutas.get("arroz", "no disponible")

print(f"Precio del pan: \t{pan}\nPrecio del arroz: \t{arroz}")
```

12. Dada la lista ["a", "b", "a", "c", "b", "a"], usa un diccionario para contar cuántas veces aparece cada elemento. Usa .get() para no romper con KeyError.

    Restricción: solo .get() para acceder al dict y for para recorrerlo · sin Counter

```python
letras = ["a", "b", "a", "c", "b", "a"]
conteo = {}

for l in letras:
  conteo[l] = conteo.get(l, 0) + 1
  
print(conteo)
```

### Acumuladores

17. Dada la lista [15, 18, 12, 20, 14], calcula el promedio acumulando la suma en una variable y dividiendo al final.

    Restricción: **sin sum()** · sin mean() · solo acumulador + for

```python
lista = [15, 18, 12, 20, 14]
suma = sum(lista)
cantidad = len(lista)

promedio = suma / cantidad

print(promedio)
```

```python
lista = [15, 18, 12, 20, 14]
suma = 0

for l in lista:
  suma += l
  
promedio = suma / len(lista)
print(promedio)
```

18. Dada la lista [9, 3, 17, 5, 11], encuentra el número mayor recorriéndola con for y usando una variable acumuladora.
    
    Restricción: sin max() · sin sorted() · solo acumulador

```python
lista = [9, 3, 17, 5, 11]
mayor = lista[0] # empezamos asumiendo que el primero es el mayor

for l in lista:
  if l > mayor:
    mayor = l # se actualiza si encontra uno más grande
print(mayor)
```

### Validación de input

19. Pide al usuario un número con input(). Si no es positivo, sigue pidiendo hasta que lo sea. Cuando lo ingrese correctamente, imprime "Válido: {número}".

    Restricción: usa while + int() + if · sin try/except

```python

while True:
  numero = int(input("Ingresa un numero: "))
  if numero > 0:
    print(f"Valido: {numero}")
    break
  
# RECUERDA: el input dentro del While para que pueda seguir pidiendolo hasta que coincida
```

20. Pide al usuario que escriba "si" o "no". Si escribe otra cosa, vuelve a pedir. Cuando sea válido, imprime "Elegiste: {respuesta}".

    Restricción: usa while + .lower() + in · sin try/except

```python
while True:
  respuesta = input("Si o No: ").lower()
  if respuesta == "si" or respuesta == "no":
    print(f"Elegiste: {respuesta}")
    break
  
# RECUERDA: el input dentro del While para que pueda seguir pidiendolo hasta que coincida
```

```python
while True:
  respuesta = input("Si o No: ").lower()
  if respuesta in ["si","no"]:
    print(f"Elegiste: {respuesta}")
    break
```
