6. Filtrar una lista de números
   
Pide al usuario 5 números usando un while y guárdalos en una lista. Al terminar, muestra dos listas separadas: una con los números pares y otra con los impares.

Reto: No puedes pedir los 5 números de una sola vez. Pídelos uno por uno dentro del bucle y ve agregándolos a la lista principal con append(). Al final, recorre la lista y clasifica con condicionales.

Pista de diseño: Necesitas tres listas: numeros = [], pares = [], impares = []. El bucle de captura usa un contador que va de 1 a 5. Para clasificar, usa n % 2 == 0.

```python
numeros = []
pares = []
impares = []

# Primer bucle, recolecta 5 numeros
while len(numeros) < 5:
  n = int(input("Ingresa un numero"))
  numeros.append(n)
  
# Segundo bucle, clasifica los numeros cuando ya tiene los 5 
for n in numeros:
  if n % 2 == 0:
    pares.append(n)
  else:
    impares.append(n)

print(f"Numeros: {numeros}")
print(f"Pares: {pares}")
print(f"Impares: {impares}")
```

7. Buscar en una lista con función

Crea una función buscar(lista, elemento) que recorra la lista con un for y retorne True si el elemento está, o False si no. Pruébala con una lista de nombres predefinida.

Reto: No puedes usar el operador in directamente — el objetivo es simular lo que hace in por dentro, con un bucle manual. Si encuentras el elemento, retorna True inmediatamente con return.

Pista de diseño: Dentro del for, compara cada elemento con == . Si hay coincidencia, return True. Si el bucle termina sin encontrarlo, return False fuera del bucle.

```python
nombres = ["Lian", "Jode", "Anlu", "Loan"]

def buscar(lista, elemento):
  for item in lista:
    if item == elemento:
      return True
  return False

def main():
  nombre = input("Bucar: ").strip().title()
  if buscar(nombres, nombre):
    print(f"\"{nombre}\" esta en la lista :)")
  else:
    print(f"\"{nombre}\" no esta en la lista :(")
    
main()
```

8. Diccionario con función — Traductor simple

Crea un diccionario con 8 palabras en español y su traducción al inglés. Luego crea una función traducir(palabra) que busque la palabra en el diccionario y retorne la traducción, o "Palabra no encontrada" si no existe.

Reto: La función no debe recibir el diccionario como parámetro — debe acceder al diccionario que está definido fuera de ella. Pruébala en un while True que siga pidiendo palabras hasta que el usuario escriba "salir".

Pista de diseño: Usa .get(clave, "Palabra no encontrada") para evitar KeyError. Normaliza la entrada con .lower().strip() antes de buscar.

```python
diccionario = {
    "hola":    "hello",
    "adiós":   "goodbye",
    "gracias": "thank you",
    "por favor": "please",
    "sí":      "yes",
    "no":      "no",
    "agua":    "water",
    "casa":    "house"
}

def traducir(palabra):
  palabra = palabra.lower().strip()
  # De "diccionario" obtenemos "palabra"
  return diccionario.get(palabra, "Palabra no encontrada")

def main():
  while True:
    palabra = input("Ingrese la palabra (o Salir): ")
    if palabra.lower() == "salir":
      break
    else:
      respuesta = traducir(palabra)
      print(f"{palabra} --> {respuesta}")
      
main()
```

9. Lista de compras con presupuesto

Crea un programa que pida productos y precios uno por uno. Guarda cada par en un diccionario dentro de una lista. El bucle termina cuando el usuario escribe "listo". Al final muestra todos los productos, el total y si supera un presupuesto de $50.

Reto: Cada elemento de la lista es un diccionario {"producto": nombre, "precio": precio}. Para calcular el total, recorre la lista y acumula los precios con un contador.

Pista de diseño: La estructura es compras = [] donde cada elemento se agrega así: compras.append({"producto": nombre, "precio": float(precio)}). Para el total: for item in compras: total += item["precio"].

```python
def main():
  compras = []
  presupuesto = 50.0
  
  # 1: Obtenemos los productos (no sabemos cuantos --> While True)
  while True:
    nombre = input("Producto (o \"listo\"): ").strip()
    if nombre.lower() == "listo":
      break
    precio = float(input(f"Precio de {nombre}: "))
    # Cada producto es un diccionario con dos campos
    compras.append({"producto": nombre, "precio": precio})
  
  # 2: Mostrar y calcular (ya sabemos cuantos hay --> for)
  print("\n--- Mi Lista ---")
  total = 0
  for item in compras:
    print(f"{item['producto']} : ${item['precio']:.2f}")
    total += item['precio'] # Acumulacion del total
    
  print(f"\nTotal: ${total:.2f}")
  
  # 3: Descision final
  if total > presupuesto:
    print(f"Superaste el presupuesto de ${presupuesto:.2f}")
  else:
    print(f"Dentro del presupuesto de ${presupuesto:.2f}")
    
main()
```

10. Función con tuplas — Conversor de unidades

Crea una lista de tuplas donde cada tupla es (valor, unidad_origen). Luego crea una función convertir(valor, unidad) que convierta: km→millas, kg→libras, °C→°F. Recorre la lista y muestra el resultado de cada conversión.

Reto: La función debe manejar las 3 conversiones con if/elif. Si la unidad no es reconocida, retorna "Unidad desconocida". Recorre la lista con un for y llama a convertir() por cada tupla usando unpacking.

Pista de diseño: Fórmulas: km × 0.621 para millas, kg × 2.205 para libras, (°C × 9/5) + 32 para °F. El unpacking en el for es: for valor, unidad in medidas:.

```python
# Una lista con tuplas
medidas = [
    (10,  "km"),
    (5,   "kg"),
    (100, "C"),
    (25,  "km"),
    (80,  "kg"),
    (37,  "C"),
]

def convertir(valor, unidad):
  if unidad == "km":
    resultado = valor * 0.621
    return f"{valor} km --> {resultado:.2f} millas"
  elif unidad == "kg":
    resultado = valor * 2.205
    return f"{valor} kg --> {resultado:.2f} libras"
  elif unidad == "C":
    resultado = (valor * 9 / 5) + 32
    return f"{valor} °C --> {resultado:.2f} °F"
  else:
    return "Unidad desconocida"
  
def main():
  for valor, unidad in medidas:
    print(convertir(valor, unidad))
    
main()
```

3. Agenda de contactos con diccionario

Crea una agenda donde cada contacto tiene nombre, teléfono y email. El usuario puede agregar contactos, buscar por nombre y listar todos. Usa un diccionario de diccionarios.

🎯 Reto
La estructura principal es agenda = {} donde cada clave es el nombre y el valor es otro diccionario con {"tel": ..., "email": ...}. Usa .get() para buscar sin errores.

💡 Pista de diseño
Para agregar: agenda[nombre] = {"tel": tel, "email": email}. Para buscar: contacto = agenda.get(nombre) — si es None, no existe. Para listar: recorre agenda.items().

```python
agenda = {}

def agregar(agenda):
  nombre = input("Nombre: ").strip().title()
  tel = input("Telefono").strip()
  email = input("Email: ").strip().lower()
  # Asignamos un diccionario como valor de la clave "nombre"
  agenda[nombre] = {"tel": tel, "email": email}
  print(f"\t\n{nombre} agregado\n")
  
def buscar(agenda, nombre):
  contacto = agenda.get(nombre)
  if contacto:
    print(f"\n\tTel: {contacto['tel']}")
    print(f"\n\tEmail: {contacto['email']}")
  else:
    print("\n\tContacto no encontrado")
    
def listar(agenda):
  if len(agenda) == 0:
    print("\n\tLa agenda esta vacia\n")
    return
  for nombre, datos in agenda.items():
    print(f"{nombre}: {datos['tel']} | {datos['email']}")

def main():
  while True:
    print("1. Agregar \n2. Buscar \n3. Listar \n4. Salir\n")
    opcion = input("Opcion: ")
    
    if opcion == "1":
      agregar(agenda)
    elif opcion == "2":
      nombre = input("Buscar nombre: ").strip().lower()
      buscar(agenda,nombre)
    elif opcion == "3":
      listar(agenda)
    elif opcion == "4":
      break

main()
```

4. Estadísticas de notas con tuplas
   
Crea una función que reciba el nombre de un alumno y una tupla de notas. Calcula y muestra el promedio, nota máxima, nota mínima y si aprobó. Pruébala con al menos 3 alumnos almacenados en una lista.

🎯 Reto
Usa una lista de tuplas donde cada tupla es (nombre, notas_tupla). La función analizar(nombre, notas) calcula las estadísticas. Recorre la lista con un for y llama a la función por cada alumno.

💡 Pista de diseño
Usa sum(notas) / len(notas) para el promedio. max(notas) y min(notas) para los extremos. El promedio >= 60 determina si aprobó.

```python
alumnos = [
  ("Ana", (85,90,78,92)),
  ("Lian", (55,60,58,62)),
  ("Carlos", (95,88,97,91))
]

def analizar(nombre, notas):
  promedio = sum(notas) / len(notas)
  maximo = max(notas)
  minimo = min(notas)
  estado = "Aprobado" if promedio >= 60 else "Desaporbado"
  
  print(f"\n--> {nombre}")
  print(f"\tNotas: {notas}")
  print(f"\tPromedio: {promedio}")
  print(f"\tMaxima: {maximo}")
  print(f"\tMinima: {minimo}")
  print(f"\tEstado: {estado}")
  
def main():
  for nombre, notas in alumnos:
    analizar(nombre, notas)
    
main()
```
