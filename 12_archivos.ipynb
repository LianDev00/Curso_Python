# Lectura y Escritura de Archivos

Hasta ahora, todo lo que el usuario ingresa se pierde cuando el programa termina.
Los **archivos** permiten **guardar datos de forma permanente**.

## Modos de apertura

| Modo | Qué hace |
|---|---|
| `"r"` | Leer (el archivo debe existir) |
| `"w"` | Escribir — **borra el contenido anterior** |
| `"a"` | Agregar al final sin borrar |
| `"r+"` | Leer y escribir |

> **Importante:** `"w"` destruye el contenido existente. Si quieres agregar, usa `"a"`.

## La forma correcta: `with open()`

```python
with open("archivo.txt", "w") as f:
    f.write("Hola mundo\n")
```

El `with` cierra el archivo automáticamente al terminar el bloque, aunque haya un error.
Siempre usa esta forma en vez de `open()` + `close()` manual.

````python
# Escribir en un archivo
with open("ejemplo.txt", "w", encoding="utf-8") as f:
    f.write("Primera línea\n")
    f.write("Segunda línea\n")

print("Archivo creado.")
````

## Leer archivos

| Método | Qué devuelve |
|---|---|
| `f.read()` | Todo el contenido como un string |
| `f.readlines()` | Lista de strings, una por línea (incluye `\n`) |
| `for linea in f:` | Itera línea por línea (eficiente para archivos grandes) |

````python
# Leer todo el archivo
with open("ejemplo.txt", "r", encoding="utf-8") as f:
    contenido = f.read()
print(contenido)

# Leer línea por línea
with open("ejemplo.txt", "r", encoding="utf-8") as f:
    for i, linea in enumerate(f, 1):
        print(f"{i}: {linea.strip()}")
````

## Manejo de errores con archivos

Si intentas abrir un archivo que no existe en modo `"r"`, Python lanza `FileNotFoundError`.
Siempre manéjalo:

````python
try:
    with open("no_existe.txt", "r", encoding="utf-8") as f:
        contenido = f.read()
except FileNotFoundError:
    print("El archivo no existe")
````

## Agregar sin borrar: modo `"a"`

````python
# Agrega al final del archivo sin borrar lo anterior
with open("ejemplo.txt", "a", encoding="utf-8") as f:
    f.write("Tercera línea agregada\n")

# Verificar
with open("ejemplo.txt", "r", encoding="utf-8") as f:
    print(f.read())
````
