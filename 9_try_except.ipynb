# Manejo de Errores: `try / except`

Hasta ahora, si el usuario escribía algo incorrecto, el programa explotaba con un error rojo. `try/except` permite **atrapar ese error** y decidir qué hacer en vez de que el programa se detenga.

## ¿Por qué necesitamos esto?

Recuerda el ejercicio 19 del repaso:

```python
numero = int(input("Ingresa un numero: "))  # si el usuario escribe "abc" → CRASH
```

El error que aparecía era:
```
ValueError: invalid literal for int() with base 10: 'abc'
```

`try/except` permite **capturar ese crash** y manejarlo con gracia.

## Estructura básica

```python
try:
    # código que puede fallar
except NombreDelError:
    # qué hacer si falla
```

Python intenta ejecutar el bloque `try`. Si ocurre un error del tipo indicado en `except`, en vez de explotar, ejecuta el bloque `except`.

````python
# Sin try/except → explota
numero = int("abc")  # ValueError
````

````python
# Con try/except → controlado
try:
    numero = int("abc")
except ValueError:
    print("Eso no es un numero valido")
````

## Errores más comunes en Python

| Error | Cuándo ocurre |
|---|---|
| `ValueError` | Convertir un valor inválido: `int("abc")` |
| `TypeError` | Operar tipos incompatibles: `"2" + 2` |
| `ZeroDivisionError` | Dividir entre cero: `10 / 0` |
| `KeyError` | Acceder a una clave que no existe en un dict: `d["clave"]` |
| `IndexError` | Acceder a un índice fuera de rango: `lista[99]` |
| `FileNotFoundError` | Abrir un archivo que no existe |

````python
# Probando cada error

# ValueError
try:
    int("hola")
except ValueError:
    print("ValueError: no se puede convertir 'hola' a int")

# ZeroDivisionError
try:
    resultado = 10 / 0
except ZeroDivisionError:
    print("ZeroDivisionError: no se divide entre cero")

# KeyError
try:
    d = {"a": 1}
    print(d["b"])
except KeyError:
    print("KeyError: la clave no existe en el diccionario")

# IndexError
try:
    lista = [1, 2, 3]
    print(lista[10])
except IndexError:
    print("IndexError: indice fuera de rango")
````

## Múltiples `except`

Puedes capturar distintos errores con bloques `except` separados:

````python
def dividir(a, b):
    try:
        resultado = int(a) / int(b)
        return resultado
    except ValueError:
        return "Error: ambos valores deben ser numeros"
    except ZeroDivisionError:
        return "Error: no se puede dividir entre cero"

print(dividir(10, 2))     # ok
print(dividir(10, 0))     # ZeroDivisionError
print(dividir(10, "abc")) # ValueError
````

## `else` y `finally`

- **`else`**: se ejecuta solo si **no** hubo error en el `try`
- **`finally`**: se ejecuta **siempre**, haya error o no (útil para cerrar archivos, limpiar recursos)

```python
try:
    # código que puede fallar
except Error:
    # si falla
else:
    # si NO falla
finally:
    # siempre se ejecuta
```

````python
# Ejemplo de else y finally
def convertir(texto):
    try:
        numero = int(texto)
    except ValueError:
        print(f"'{texto}' no es un numero")
    else:
        print(f"Conversion exitosa: {numero}")
    finally:
        print("--- fin del intento ---")

convertir("42")
print()
convertir("abc")
````

## Patrón más usado: validar input en un bucle

Este es el patrón que más vas a usar — combina `while True` con `try/except` para seguir pidiendo hasta que el usuario ingrese algo válido:

````python
while True:
    try:
        numero = int(input("Ingresa un numero entero: "))
        break  # solo sale si int() tuvo exito
    except ValueError:
        print("Eso no es un entero, intenta de nuevo")

print(f"Ingresaste: {numero}")
````

### 1. Conversion segura

Pide un numero al usuario y conviertelo a `int`. Si no es valido, imprime `"Error: ingresa un numero entero"`.

**Reto:** No uses `while` — solo un intento. Si falla, termina con el mensaje de error.

### 2. Division sin crash

Pide dos numeros al usuario y divide el primero entre el segundo. Maneja dos posibles errores: que no sean numeros y que el divisor sea cero.

**Reto:** Usa dos bloques `except` separados, uno por cada tipo de error.

````python
try:
  a = int(input("a: "))
  b = int(input("b: "))
  division = a / b
  print(f"{division:.0f}")
except ValueError:
  print("Error: ingresa un numero entero")
except ZeroDivisionError:
  print("Error: no se puede dividir por 0")
````

### 3. Acceso seguro a lista

Dada la lista `[10, 20, 30, 40, 50]`, pide al usuario un indice y muestra el elemento. Si el indice esta fuera de rango, muestra `"Indice invalido"`.

### 4. Busqueda segura en diccionario

Dado el diccionario `{"peru": "lima", "chile": "santiago", "argentina": "buenos aires"}`, pide un pais y muestra su capital. Si no existe, muestra `"Pais no encontrado"`.

**Reto:** Usa `except KeyError` en vez de `.get()` — ambos resuelven el mismo problema pero de forma distinta.

**Pista:** Accede con `capitales[pais]` dentro del `try`. Si la clave no existe, lanza `KeyError`.

````python
capitales = {"peru": "lima", "chile": "santiago", "argentina": "buenos aires"}

try:
  pais = input("Ingresa un pais:")
  print(capitales[pais])
except KeyError:
  print("Pais no encontrado")
````

### 5. Input numerico con else

Pide un numero al usuario. Si es valido, usa el bloque `else` para imprimir `"Numero valido: {n}"`. Si no, imprime `"No es un numero"`. Usa `finally` para imprimir `"Intento finalizado"` siempre.

**Reto:** Usa los tres bloques: `except`, `else` y `finally`.

**Pista:** `else` solo se ejecuta si el `try` no lanza ningun error.

````python
try:
  numero = int(input("Ingresa un numero: "))
except ValueError:
  print("No es un numero")
else:
  print(f"Numero valido: {numero}")
finally:
  print("---- Intento finalizado ----")
````

### 6. Input validado con bucle

Crea una funcion `pedir_entero(mensaje)` que siga pidiendo al usuario hasta que ingrese un numero entero valido. Cuando lo ingrese, retorna el numero.

Luego usala para pedir edad y mostrar `"Tu edad es: {edad}"`.

**Reto:** La funcion no imprime el resultado — solo valida y retorna. El print va en `main()`.

**Pista:** Combina `while True` + `try/except` + `break` dentro de la funcion. Si el `int()` falla, imprime el error y el bucle vuelve a pedir.

**Pasos:**
- Paso 1: define `pedir_entero(mensaje)` con `while True` adentro
- Paso 2: dentro del `while`, usa `try` para convertir el input a `int`
- Paso 3: si tiene exito, `return` el numero (eso tambien sale del bucle)
- Paso 4: si falla, el `except` imprime el error y el bucle repite
- Paso 5: en `main()`, llama a la funcion y usa el resultado

````python
def pedir_entero(mensaje):
  while True:
    try:
      numero = int(input(mensaje))
    except ValueError:
      print("Error: se debe ingresar un entero")
    else:
      return numero
    
def main():
  mensaje = pedir_entero(numero)
  print(f"Tu edad es: {mensaje}")
  
main()
````

### 7. Calculadora con menu

Crea una calculadora con menu que soporte suma, resta, multiplicacion y division. El menu se repite con `while True` hasta que el usuario elija salir.

Maneja todos los posibles errores: entrada no numerica, division entre cero, opcion invalida.

**Reto:** Crea una funcion `pedir_numero(mensaje)` reutilizable que valide el input, y llamala dos veces por operacion.

**Pista:** El menu acepta opciones `"1"`, `"2"`, `"3"`, `"4"`, `"5"` (salir). Usa `elif` para cada operacion y un `else` para opcion invalida.

**Pasos:**
- Paso 1: define `pedir_numero(mensaje)` con `while True` + `try/except ValueError`
- Paso 2: el bucle principal del menu es otro `while True`
- Paso 3: pide la opcion y usa `if/elif/else` para cada caso
- Paso 4: para la division, agrega `except ZeroDivisionError` al hacer la operacion
- Paso 5: la opcion de salir usa `break`

````python
def pedir_numero(mensaje):
  while True:
    try:
      return float(input(mensaje))
    except ValueError:
      print("Error: ingresa un numero valido")
      
def main():
  while True:
    opcion = input("1.Suma \n2.Resta \n3.Multiplicacion \n4.Division \n5.Salir")
    if opcion == "5":
      print("--- Salir... ---")
      break
    elif opcion in ["1","2","3","4"]:
      a = pedir_numero("Numero a: ")
      b = pedir_numero("Numero b: ")
      if opcion == "1":
        print(f"Suma: {a + b}")
      elif opcion == "2":
        print(f"Resta: {a - b}")
      elif opcion == "3":
        print(f"Multiplicacion: {a * b}")
      elif opcion == "4":
        try:
          print(f"Division: {a / b}")
        except ZeroDivisionError:
          print("Error: no se puede dividir entre cero")
      else:
        print("--- Opcion invalida ---")

main()
````
