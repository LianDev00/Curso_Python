3. Cajero Automático

Simula un cajero con saldo inicial de $1000. El usuario puede retirar dinero. El programa repite hasta que el saldo llegue a $0 o el usuario escriba "salir".

Reto: Valida que el monto a retirar no sea mayor al saldo disponible ni negativo. Muestra el saldo restante después de cada retiro.

Pista de diseño: Usa while True con dos condiciones de salida: si el usuario escribe "salir" → break, y si el saldo llega a 0 → break. Convierte la entrada a int solo si no es "salir".

```python
# Mi solucion
saldo = 1000
print(F"Saldo Total: {saldo}")

while True:
  entrada = input("Saldo a retirar (o Salir): ")
  if entrada == "salir" or saldo == 0:
    print("--- Saldo agotado ---")
    break
  saldo = saldo - int(entrada)
  print(F"Saldo Actual: {saldo}")
  
print(f"Saldo Total: {saldo}")
```

Ej. 3 — Cajero 

Funciona, pero tiene un bug: si retiras más de lo que tienes, el saldo se vuelve negativo sin avisar. Falta validar que int(entrada) <= saldo.

```python
# Solucion de Claude
saldo = 1000
print(f"Saldo inicial: ${saldo}")

while True:
  entrada = input("Monto a retirar (o \"Salir\"): ").strip().lower()
  
  if entrada == "salir":
    break

  monto = int(entrada)
  
  if monto <= 0:
    print("--- Monto Invalida ---")
  elif monto > saldo:
    print(f"Saldo insuficiente. Tienes ${saldo}")
  else:
    saldo -= monto
    print(f"Saldo restante: ${saldo}")
  
  if saldo == 0:
    print("--- Saldo Agotado ---")
    break

print(f"Saldo Final: ${saldo}")
```

4. Suma hasta que el Usuario Diga Basta

Pide números al usuario uno por uno y los va sumando. El bucle termina cuando el usuario ingresa 0. Al final muestra la suma total y cuántos números ingresó.

Reto: Usa break cuando detectes el 0. No cuentes el 0 en el total de números.

Pista de diseño: Lleva dos contadores: uno para la suma (total) y otro para la cantidad de números (cantidad). Actualiza ambos dentro del while, pero solo si el número no es 0.

```python
total = 0
cantidad = 0

while True:
  numero = int(input("Ingresa un numero (0 para terminar): "))
  
  if numero == 0:
    break
  
  total += numero # Acumular suma
  cantidad += 1 # Contar cuantos numeros ingreso
  
if cantidad > 0: # Evitar division por cero
  promedio = total / cantidad
  print(f"Numero ingresados: {cantidad}")
  print(f"Suma Total: {total}")
  print(f"Promedio: {promedio:.2f}")
else:
  print("No ingresaste ningun numero")
```
