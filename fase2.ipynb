1. Analizador de lista de numeros

    Dada la lista [12, 7, 25, 3, 18, 9, 14], crea un programa que:
    - Calcule e imprima el promedio
    - Imprima cuántos números están por encima del promedio
    - Imprima cuántos están por debajo o igual

    Ejemplo de salida:
    - Promedio: 12.57
    - Por encima: 3
    - Por debajo o igual: 4

    Pasos:
    - Paso 1: calcula el promedio (igual que ejercicio #17)
    - Paso 2: recorre la lista con for (igual que ejercicio #01)
    - Paso 3: por cada elemento, pregunta si supera el promedio (igual que ejercicio #05)
    - Paso 4: acumula los conteos en dos variables separadas

````python
lista = [12, 7, 25, 3, 18, 9, 14]
promedio = sum(lista) / len(lista)
conteo_max = []
conteo_min = []

print(f"Promedio: {promedio:.2f}\n")

for l in lista:
  if l > promedio:
    conteo_max.append(l)
  else:
    conteo_min.append(l)

print(f"Por encima del promedio: {conteo_max}  --> {len(conteo_max)}")
print(f"Por debajo del promedio: {conteo_min} --> {len(conteo_min)}")
````

2. Registro de palabras unicas

    Pide al usuario palabras una por una. Si escribe "fin", termina.
    Al terminar, muestra:
    - La lista completa de palabras ingresadas
    - Las palabras únicas (sin repetir)
    - Cuántas veces se repitió alguna palabra

    Ejemplo:
    - Ingresa: casa / árbol / casa / mar / árbol / fin
    - Todas: ['casa', 'árbol', 'casa', 'mar', 'árbol']
    - Únicas: {'árbol', 'casa', 'mar'}
    - Repetidas: 2

    Pasos:
      - Paso 1: valida el input en un while (ejercicio #19 — pero la condición de salida es "fin")
      - Paso 2: agrega cada palabra a una lista con append (ejercicio #09)
      - Paso 3: convierte la lista a set para obtener únicas (ejercicio #13)
      - Paso 4: la diferencia de tamaños entre las listas te dan las repeticiones

````python
lista = []
while True:
  palabras = input("Ingresa palabras: ")
  if palabras == "fin":
    break
  lista.append(palabras)
  new_lista = sorted(list(set(lista)))
  repetidas = len(lista) - len(new_lista)
  
print(lista)
print(new_lista)
print(f"Repetidas: {repetidas}")
  
````

3. Calculadora de calificaciones

    Dada la lista de notas [15, 18, 9, 12, 20, 7, 14, 11]:
    
    - Crea una función que reciba una nota y retorne su categoría:
      - 0–10 → "Desaprobado" 
      - 11–14 → "Regular" 
      - 15–17 → "Bueno" 
      - 18–20 → "Excelente"
    - Aplica la función a cada nota
    - Cuenta cuántos hay en cada categoría (usa diccionario)
    - Imprime el resumen y el promedio general
    
    Pasos:
    - Paso 1: define la función clasificar(nota) con if/elif/else y return (ejercicios #07 + #06)
    - Paso 2: recorre la lista con for, llama la función por cada nota
    - Paso 3: acumula en un dict usando .get() (ejercicio #12)
    - Paso 4: calcula el promedio con acumulador (ejercicio #17)

````python
notas = [15, 18, 9, 12, 20, 7, 14, 11]
conteo = {}

def clasificar(nota):
  if nota >= 18:
    return "Excelente"
  elif nota >= 15:
    return "Bueno "
  elif nota >= 11:
    return "Regular"
  else:
    return "Desaprobado"

suma = 0

for n in notas:
  # Clasificamos cada nota n usando la función clasificar y la guardamos en "categoria". 
  categoria = clasificar(n)
  # Luego obtenemos cuántas notas tiene esa categoría en el diccionario conteo y la guardamos en "actual". 
  actual = conteo.get(categoria, 0)
  # Finalmente, le sumamos 1 y actualizamos el diccionario — si la categoría no existía, empieza en 1.
  conteo[categoria] = actual + 1
  suma += n
  
promedio = suma / len(notas)

print(f"PROMEDIO: \t{promedio:.2f}")
for categoria, cantidad in conteo.items():
  print(f"{categoria}: \t{cantidad}")
````

4. Mini agenda de contactos

    Crea una mini agenda que funcione en un menú de consola:

    Opciones: "agregar", "buscar", "salir"

    - agregar: pide nombre y teléfono, los guarda en un diccionario
    - buscar: pide un nombre, lo busca con .get() y muestra "Hola, {nombre}: {tel}" o "No encontrado"
    - salir: termina el programa

    El bucle while continúa hasta que el usuario escriba "salir". Valida que la opción sea una de las tres válidas.

    Pasos:
      - Paso 1: el bucle principal es un while (ejercicio #03) con condición de salida "salir"
      - Paso 2: valida que la opción ingresada sea válida (ejercicio #20 — in con opciones)
      - Paso 3: el dict almacena contactos, accedes con .get() (ejercicio #11)
      - Paso 4: el mensaje de búsqueda usa f-string (ejercicio #08)
      - Paso 5: usa .lower() en el nombre para no distinguir mayúsculas (ejercicio #15)

````python
contactos = {
  "Lian": "999-111",
  "Mili": "999-222",
  "Anlu": "999-333"
}

while True:
  opcion = input("1. Agregar \n2. Buscar \n3. Salir ").strip()
  if opcion == "3":
    break
  if opcion == "1":
    nombre = input("Agregar contacto: ").strip().title()
    tel = input("Agregar numero: ").strip()
    contactos[nombre] = tel
  if opcion == "2":
    nombre = input("Buscar contacto: ").strip().title()
    # Buscamos el telefono segun el nombre y lo guardamos en la variable "tel", "get(clave) = valor"
    tel = contactos.get(nombre, "No encontrado")
    print(f"Hola, {nombre}: {tel}")
  
````

5. Analizador de texto completo

    Dado el string:
    "el sol sale siempre sobre el mismo suelo"
    
    Crea un programa que muestre:
    - Total de vocales en el texto completo
    - Frecuencia de cada palabra (cuántas veces aparece cada una)
    - La palabra que más se repite
    - Lista de palabras con más de 4 letras
    - Total de palabras en el texto
  
    Pasos:
      - Paso 1: recorre el string letra por letra para contar vocales (ejercicio #16)
      - Paso 2: usa .split() para separar palabras, luego for + dict para frecuencias (ejercicio #12)
      - Paso 3: recorre el dict para encontrar la palabra con más ocurrencias — acumulador de máximo      (ejercicio #18)
      - Paso 4: filtra palabras largas con for + if + append (ejercicio #10)
      - Paso 5: cuenta total de palabras con acumulador en for (ejercicio #02)

````python
string = "el sol sale siempre sobre el mismo suelo"

vocal = 0
for v in string:
  if v in "aeiou":
    vocal += 1

palabras = {}
for p in string.split():
  palabras[p] = palabras.get(p, 0) + 1

p_larga = []
for pl in string.split():
  if len(pl) > 4:
    p_larga.append(pl)

p_max = ""
conteo_max = 0
for pt, conteo in palabras.items():
  if conteo > conteo_max:
    conteo_max = conteo
    p_max = pt

total = len(string.split())
  
print(f"Vocales: {vocal}")
print(f"Frecuencia: {palabras}")
print(f"Mas repetida: \"{p_max}\" ({conteo_max} veces)")
print(f"Palabras largas: {p_larga}")
print(f"Total de palabras: {total}")
````

### Ejercicio A — Contador de colores

Dada la lista `["rojo", "azul", "rojo", "verde", "azul", "rojo", "verde", "azul"]`:

- Usa un diccionario para contar cuántas veces aparece cada color
- Luego recorre el diccionario y encuentra el color que más se repite
- Imprime el conteo completo y el color ganador

Ejemplo de salida:
```
rojo: 3
azul: 3
verde: 2
Más repetido: rojo
```

Restricciones: usa `.get()` para el conteo · sin `Counter` · encuentra el máximo con acumulador en `for` · sin `max()`

````python
colores = ["rojo", "azul", "rojo", "verde", "azul", "rojo", "verde", "azul"]
conteo = {}

for c in colores:
  conteo[c] = conteo.get(c, 0) + 1
print(conteo)

color_max = ""
max_conteo = 0
for color, cantidad in conteo.items():
  if cantidad > max_conteo:
    max_conteo = cantidad
    color_max = color
print(f"Mas repedito: {color_max}")
````

### Ejercicio B — Inventario de tienda

Tienes este inventario:
```python
inventario = {
    "manzana": {"precio": 2.50, "stock": 10},
    "pera":    {"precio": 3.00, "stock": 5},
    "uva":     {"precio": 5.50, "stock": 8},
}
```

- Muestra todos los productos con su precio y stock
- Calcula el valor total del inventario (precio × stock de cada producto, sumados)
- Pide al usuario un producto y una cantidad. Si hay stock suficiente, descuéntalo e imprime `"Compra exitosa"`. Si no hay suficiente, imprime `"Stock insuficiente"`
- Usa `.get()` para buscar el producto — si no existe, imprime `"Producto no encontrado"`

Ejemplo de salida:
```
manzana: $2.50 | stock: 10
pera: $3.00 | stock: 5
uva: $5.50 | stock: 8
Valor total del inventario: $94.00
Producto: uva
Cantidad: 3
Compra exitosa. Stock restante de uva: 5
```

Restricciones: usa `.get()` para buscar · accede a los sub-diccionarios con `producto["precio"]` · sin funciones propias

Pasos:
- Paso 1: recorre `inventario.items()` con `for` para mostrar cada producto — cada valor es un dict con `"precio"` y `"stock"`
- Paso 2: en ese mismo `for`, acumula `precio × stock` en una variable `total` para el valor del inventario
- Paso 3: pide el nombre del producto con `input()` y búscalo con `.get()` — guarda el resultado en una variable (será `None` si no existe)
- Paso 4: si el resultado es `None` → "Producto no encontrado". Si existe, pide la cantidad
- Paso 5: compara la cantidad con `producto["stock"]` — si hay suficiente, resta y actualiza. Si no, avisa

````python
inventario = {
    "manzana": {"precio": 2.50, "stock": 10},
    "pera":    {"precio": 3.00, "stock": 5},
    "uva":     {"precio": 5.50, "stock": 8},
}
total = 0

for nombre, datos in inventario.items():
  precio = datos["precio"]
  stock = datos["stock"]
  print(f"{nombre}: ${precio} | stock: {stock}")
  total += precio * stock
  
print(f"Valor total: {total:.2f}")

buscar = input("Producto: ")
producto = inventario.get(buscar)

if producto is None:
  print("Producto no encontrado")
else:
  cantidad = int(input("Cantidad: "))
  if cantidad <= producto["stock"]:
    producto["stock"] -= cantidad
    print(f"Compra exitosa. Stock restante de \"{buscar}\": {producto["stock"]}")
  else:
    print("Stock insuficiente")
````
