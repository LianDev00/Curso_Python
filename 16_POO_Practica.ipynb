### 4. `EstadisticasNumeros` (inicialización correcta)

⚠️ Revisita el **bug del ejercicio 8**: si inicializas `minimo = 0`, el programa siempre devuelve 0 cuando todos los números son positivos.

Crea una clase `EstadisticasNumeros` con:
- `__init__(self, numeros)` → guarda la lista
- `minimo()`, `maximo()`, `promedio()` → sin usar `min`, `max`, `sum`
- Si la lista está vacía → `raise ValueError("Lista vacía")`

Prueba con `[8, 3, 15, 22, 4]` — el mínimo debe ser `3`, NO `0`.

**Reto:** Inicializa `mn = self.numeros[0]` y `mx = self.numeros[0]` (con el primer elemento), luego recorre desde el segundo en adelante.

**Pista:**
```python
mn = self.numeros[0]
for n in self.numeros[1:]:
    if n < mn:
        mn = n
```
