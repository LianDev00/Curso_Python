```python
camel = input("Camel: ")
snake = ""
for c in camel:
  if c.isupper():
    snake += "_" + c.lower()
  else:
    snake += c

print(f"camelCase: {camel}")
print(f"snake_case: {snake}")
```

```python
monto = 0
while monto < 50:
  print(f"Monto pendiene: {50 - monto}")
  moneda = int(input("Moneda: "))
  if moneda in [25,10, 5]:
    monto += moneda
  
print(f"Cambio: {monto - 50}")
```

```python
def is_valid(s):
    # La placa debe tener entre 2 y 6 caracteres
    if len(s) < 2 or len(s) > 6:
        return False

    # Los primeros 2 caracteres deben ser letras obligatoriamente
    if not s[:2].isalpha():
        return False

    # Solo se permiten letras y números (sin espacios, puntos, etc.)
    if not s.isalnum():
        return False

    # Rastreamos si ya encontramos algún dígito
    found_digit = False

    # Recorremos cada carácter para detectar letras después de números
    for c in s:
        if c.isdigit():
            found_digit = True
        elif found_digit:
            # Si ya apareció un dígito y ahora hay una letra → inválido
            return False

    # Buscamos el primer dígito de la placa
    for i, c in enumerate(s):
        if c.isdigit():
            # El primer dígito no puede ser 0
            if c == "0":
                return False
            break  # Solo nos importa el primero, salimos del loop

    return True


def main():
    plate = input("Plate: ")
    print("Valid" if is_valid(plate) else "Invalid")


main()
```
