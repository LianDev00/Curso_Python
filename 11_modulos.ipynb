# Módulos Estándar: `random`, `datetime`, `os`

Python viene con muchas **librerías incluidas** que no necesitas instalar.
Solo las importas y ya puedes usarlas.

## ¿Qué es un módulo?

Un módulo es un archivo Python con funciones y variables listas para usar.
Se importa con la palabra clave `import`:

```python
import random          # importa todo el módulo
from random import randint  # importa solo una función
```

## `random` — números y elecciones aleatorias

| Función | Qué hace |
|---|---|
| `random.randint(a, b)` | Entero aleatorio entre `a` y `b` (ambos incluidos) |
| `random.random()` | Float aleatorio entre 0.0 y 1.0 |
| `random.choice(lista)` | Elige un elemento aleatorio de la lista |
| `random.shuffle(lista)` | Mezcla la lista en su lugar (modifica la original) |
| `random.sample(lista, k)` | Elige `k` elementos únicos aleatorios (no modifica la original) |

````python
import random

print(random.randint(1, 10))         # ej: 7
print(random.choice(["rojo", "azul", "verde"]))  # ej: 'azul'

cartas = ["A", "2", "3", "J", "Q", "K"]
random.shuffle(cartas)
print(cartas)  # orden aleatorio

print(random.sample(range(1, 50), 6))  # 6 números únicos del 1 al 49 (lotería)
````

## `datetime` — fechas y tiempos

| Función | Qué hace |
|---|---|
| `date.today()` | La fecha de hoy |
| `datetime.now()` | La fecha y hora actual |
| `timedelta(days=n)` | Una duración de `n` días |
| `.strftime(formato)` | Convierte fecha a string con formato |
| `.weekday()` | Día de la semana (0=lunes, 6=domingo) |

````python
from datetime import date, datetime, timedelta

hoy = date.today()
print(hoy)                     # 2026-04-12
print(hoy.strftime("%d/%m/%Y"))  # 12/04/2026

ahora = datetime.now()
print(ahora.strftime("%H:%M"))   # hora actual

manana = hoy + timedelta(days=1)
print(manana)

dias_semana = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]
print(dias_semana[hoy.weekday()])  # 'Sábado'
````

## `os` — sistema de archivos

| Función | Qué hace |
|---|---|
| `os.getcwd()` | Directorio actual |
| `os.listdir(ruta)` | Lista archivos en una carpeta |
| `os.path.exists(ruta)` | ¿Existe el archivo o carpeta? |
| `os.path.join(a, b)` | Une rutas de forma segura |

````python
import os

print(os.getcwd())             # directorio actual
print(os.path.exists("notas.txt"))  # True o False

ruta = os.path.join("carpeta", "archivo.txt")
print(ruta)  # carpeta/archivo.txt (o carpeta\archivo.txt en Windows)
````

### 4. Simulador de dado

Simula lanzar un dado de 6 caras **20 veces**. Cuenta cuántas veces salió cada número y muestra el resultado.

**Resultado esperado** (ejemplo):
```
1 → 3 veces
2 → 4 veces
3 → 2 veces
4 → 5 veces
5 → 3 veces
6 → 3 veces
```

**Reto:** Sin hardcodear los 6 números — el conteo debe funcionar para cualquier dado.

**Pista:** Usa el patrón contador con diccionario: `conteo[resultado] = conteo.get(resultado, 0) + 1`.

````python
import random
conteo = {}
for d in range(20):
  resultado = random.randint(1, 6)
  conteo[resultado] = conteo.get(resultado, 0) + 1

for numero, veces in sorted(conteo.items()):
  print(f"{numero} -> {veces} veces")
````

### 5. Días para Año Nuevo

Calcula cuántos días faltan para el próximo 1 de enero y muéstralo.

**Resultado esperado** (ejemplo): `Faltan 264 días para Año Nuevo.`

**Reto:** El programa debe funcionar en cualquier fecha del año, no solo hoy.

**Pista:** Crea la fecha del `1 de enero del año siguiente` con `date(hoy.year + 1, 1, 1)`. Resta las fechas y usa `.days`.

````python
from datetime import date
hoy = date.today()
primer_año = date(hoy.year + 1, 1, 1)
dias_faltan = primer_año - hoy
print(f"Faltan {dias_faltan.days} dias para Año Nuevo.")
````

## Difíciles

### 6. Adivina el número

El programa elige un número aleatorio entre 1 y 50. El usuario tiene **5 intentos** para adivinarlo.
En cada intento incorrecto, indica si el número secreto es mayor o menor.

**Reglas:**
- Si el usuario no ingresa un número entero → muestra error y no consume el intento
- Si adivina → `"¡Adivinaste en X intentos!"`
- Si se acaban los intentos → `"Perdiste. El número era: X"`

**Reto:** Usa `for` con intentos, `try/except ValueError` para el input, y `break` al adivinar.

**Pista:** El `else` del `for` se ejecuta si no hubo `break` (el usuario perdió).

````python
import random
numero = random.randint(1,50)
intentos = 5

for intento in range(1, intentos + 1):
  try:
    intent = int(input(f"Intento {intento}/{intentos}. Adivina el numero: "))
    
    if intent == numero:
      print(f"{intent}: Adivinaste en {intento} intentos!")
      break
    elif intent < numero:
      print(f"{intent}: El numero es mayor")
    else:
      print(f"{intent}: El numero es menor")
  except ValueError:
    print("Error: Ingresa un numero valido")
else:
  print(f"Perdiste. El numero era: {numero}")
````

### 7. Generador de contraseñas

Crea una función `generar_password(longitud)` que genere una contraseña aleatoria con:
- Al menos una mayúscula
- Al menos una minúscula  
- Al menos un número
- El resto rellena con caracteres aleatorios de los tres tipos
- `longitud` mínima: 8

Luego pide al usuario la longitud deseada (validada con `try/except`) y muestra la contraseña generada.

**Resultado esperado** (ejemplo): `Contraseña generada: Kp3mRx9v`

**Reto:** Garantiza que siempre haya al menos uno de cada tipo, no solo por suerte.

**Pista:**
- `import string` → `string.ascii_uppercase`, `string.ascii_lowercase`, `string.digits`
- Empieza con `[random.choice(mayus), random.choice(minus), random.choice(nums)]` para garantizar uno de cada uno
- Rellena el resto con `random.choice(todos)` y luego mezcla la lista con `random.shuffle()`
- Convierte la lista a string con `"".join(lista)`

````python
import string

def generar_password(longitud):
  mayus = string.ascii_uppercase
  minus = string.ascii_lowercase
  nums = string.digits
  simbolos = string.punctuation
  todos = mayus + minus + nums

  password = [
    random.choice(mayus), 
    random.choice(minus), 
    random.choice(nums),
    random.choice(simbolos),
  ]

  for _ in range(longitud - 4):
    password.append(random.choice(todos))
    
  random.shuffle(password)
  return "".join(password)

try:
  longitud = int(input("Longitud: "))
  if longitud < 8:
    print("Se necesitan al menos 8 caracteres")
  else:
    print(f"Password generada: {generar_password(longitud)}")
except ValueError:
  print("Debe ser un numero")
````
