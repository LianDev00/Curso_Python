# List Comprehensions

Una **list comprehension** es una forma compacta de crear listas en Python.
En vez de usar un `for` con `append`, puedes escribirlo todo en una sola línea.

## Comparación: for tradicional vs comprehension

```python
# Forma tradicional
cuadrados = []
for n in range(5):
    cuadrados.append(n ** 2)

# Con list comprehension
cuadrados = [n ** 2 for n in range(5)]
```

Ambas producen `[0, 1, 4, 9, 16]`. La segunda es más corta y legible.

## Estructura básica

```python
[expresion for variable in iterable]
```

- **expresion**: qué quieres guardar en la lista
- **variable**: el elemento de cada iteración
- **iterable**: la secuencia que recorres (lista, range, string, etc.)

````python
# Ejemplos básicos
numeros = [n for n in range(1, 6)]
print(numeros)  # [1, 2, 3, 4, 5]

dobles = [n * 2 for n in range(1, 6)]
print(dobles)  # [2, 4, 6, 8, 10]

letras = [c.upper() for c in "hola"]
print(letras)  # ['H', 'O', 'L', 'A']
````

## Con filtro: `if` al final

```python
[expresion for variable in iterable if condicion]
```

El `if` al final **filtra** — solo incluye los elementos donde la condición es `True`.

````python
# Solo los pares del 1 al 10
pares = [n for n in range(1, 11) if n % 2 == 0]
print(pares)  # [2, 4, 6, 8, 10]

# Solo palabras largas
palabras = ["sol", "lluvia", "mar", "tormenta", "rio"]
largas = [p for p in palabras if len(p) > 4]
print(largas)  # ['lluvia', 'tormenta']
````

## Dict comprehension

Mismo concepto, pero para crear diccionarios:

```python
{clave: valor for variable in iterable}
```

````python
# Palabra → su longitud
palabras = ["python", "lista", "bucle"]
longitudes = {p: len(p) for p in palabras}
print(longitudes)  # {'python': 6, 'lista': 5, 'bucle': 5}

# Número → su cuadrado
cuadrados = {n: n**2 for n in range(1, 6)}
print(cuadrados)  # {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
````

## Cuándo NO usarlas

Si la lógica es compleja (varios `if/else` anidados, múltiples líneas), un `for` tradicional es más claro.
Las comprehensions son para casos **simples y directos**.

```python
# Esto ya es demasiado — mejor un for normal
resultado = [x if x > 0 else -x for x in lista if x != 0 and x % 2 == 0]
```

# Ejercicios

## Fáciles

### 1. Cuadrados del 1 al 10

Crea una lista con los cuadrados de los números del 1 al 10 usando una list comprehension.

**Resultado esperado:** `[1, 4, 9, 16, 25, 36, 49, 64, 81, 100]`

**Reto:** Una sola línea.

**Pista:** `n ** 2` es el cuadrado de `n`.

### 2. Filtrar pares

Dada la lista `[3, 8, 1, 14, 7, 22, 5, 16]`, crea una nueva lista con solo los números pares.

**Resultado esperado:** `[8, 14, 22, 16]`

**Reto:** Usa el `if` al final de la comprehension.

**Pista:** Un número es par si `n % 2 == 0`.

### 3. Nombres en mayúscula

Dada la lista `["ana", "carlos", "beatriz", "diego"]`, crea una nueva lista con cada nombre en mayúscula usando `.upper()`.

**Resultado esperado:** `['ANA', 'CARLOS', 'BEATRIZ', 'DIEGO']`

**Reto:** Una sola línea.

**Pista:** La expresión va antes del `for`.

---
## Intermedios

### 4. Palabras largas en title case

Dada la lista `["sol", "tormenta", "mar", "relampago", "rio", "nube"]`, crea una lista con solo las palabras que tienen más de 4 letras, convertidas a `.title()`.

**Resultado esperado:** `['Tormenta', 'Relampago']`

**Reto:** Combina la expresión (`.title()`) con el filtro (`if len(...) > 4`) en una sola línea.

**Pista:** La expresión y el filtro son cosas separadas — la expresión transforma, el `if` filtra.

### 5. Dict de longitudes

Dada la lista `["python", "java", "javascript", "go", "rust"]`, crea un diccionario donde la clave es el lenguaje y el valor es su longitud.

**Resultado esperado:** `{'python': 6, 'java': 4, 'javascript': 10, 'go': 2, 'rust': 4}`

**Reto:** Usa dict comprehension `{clave: valor for ...}`.

**Pista:** La clave es el nombre del lenguaje, el valor es `len(nombre)`.

---
## Difíciles

### 6. Aplanar lista de listas

Dada la lista `[[1, 2, 3], [4, 5], [6, 7, 8, 9]]`, crea una lista plana con todos los números juntos.

**Resultado esperado:** `[1, 2, 3, 4, 5, 6, 7, 8, 9]`

**Reto:** Usa una comprehension con dos `for` — uno para la sublista y otro para el elemento.

**Pista:** `[elemento for sublista in lista for elemento in sublista]`

### 7. Tabla de multiplicar

Crea una lista de strings con la tabla de multiplicar del 1 al 5 (solo hasta el 5x5).

**Resultado esperado:**
```
['1x1=1', '1x2=2', '1x3=3', '1x4=4', '1x5=5',
 '2x1=2', '2x2=4', ..., '5x5=25']
```

**Reto:** Una sola comprehension con dos `for` anidados.

**Pista:** `f"{a}x{b}={a*b}"` es la expresión. Los dos `for` van: `for a in range(1, 6) for b in range(1, 6)`.
